```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('student_feedback.csv')
print("CSV file loaded successfully.")

# Standardize column names
df.columns = df.columns.str.replace(' ', '_')
print("\nColumn names standardized.")

print("\nDataFrame Head:")
print(df.head())

print("\nDataFrame Info:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

# Calculate average ratings
rating_columns = df.drop(columns=['Student_ID']).columns
average_ratings = df[rating_columns].mean().sort_values(ascending=False)

print("\nAverage Ratings for each feedback:")
print(average_ratings)

highest_rated_aspect = average_ratings.idxmax()
lowest_rated_aspect = average_ratings.idxmin()

print(f"\nHighest rated aspect: '{highest_rated_aspect}' with average rating {average_ratings.max():.2f}")
print(f"Lowest rated aspect: '{lowest_rated_aspect}' with average rating {average_ratings.min():.2f}")

# -------------------- BAR CHART --------------------
plt.figure(figsize=(10, 6))
plt.bar(average_ratings.index, average_ratings.values)
plt.title('Bar Chart of Average Ratings per Feedback Aspect', fontsize=16)
plt.xlabel('Feedback Aspect', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.ylim(0, 10)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
